# 🧪 Ejercicios de Exploración y Limpieza de Datos
**Dataset: 2020_Accidentalidad.xlsx (accidentes de tráfico en Madrid)**  
**Tecnología: Python + Pandas**

---
### 💡 Contexto

Este dataset contiene información sobre los accidentes de tráfico ocurridos en Madrid durante el año 2020. A lo largo de estos ejercicios vamos a analizar, limpiar y preparar los datos para un análisis más profundo.

---
### Ejercicios (sin resolver)

## 🔎 Ejercicio 1. Cargar y visualizar los datos

Carga el archivo `2020_Accidentalidad.xlsx`, hoja `2020_Accidentalidad`, parseando la columna de fecha y utilizando como índice el número de expediente. Visualiza las primeras filas del dataframe.

## 📐 Ejercicio 2. Diagnóstico inicial

1. ¿Cuántas filas y columnas tiene el dataset?
2. ¿Qué tipo de datos tiene cada variable?
3. ¿Qué porcentaje de valores nulos hay por variable?
4. ¿Cuánto ocupa el dataset en memoria?

## 🧹 Ejercicio 3. Limpieza de nombres

Renombra todas las columnas para que no tengan espacios, ni acentos, ni mayúsculas. Asegúrate de que los nombres sean válidos para trabajar cómodamente en Python.

## 🧼 Ejercicio 4. Eliminar columnas innecesarias

Elimina del dataframe las columnas que no contienen información útil.

## 🧭 Ejercicio 5. Duplicados

1. ¿Hay registros duplicados?
2. Si los hay, elimínalos.

## 🧾 Ejercicio 6. Conversión de tipos

Convierte las siguientes variables a tipo `category`:
- `distrito`
- `tipo_accidente`
- `tiempo`
- `vehiculo`
- `persona`
- `edad`
- `sexo`

Explica por qué puede ser útil esta conversión.

## 🧩 Ejercicio 7. Imputación de nulos

1. Elimina los registros con valores nulos en las columnas `numero` y `distrito`.
2. Imputa los nulos de `persona`, `vehiculo` y `tipo_accidente` utilizando la moda.
3. Añade una categoría “Desconocido” en las variables `sexo` y `tiempo` para imputar sus nulos.
4. Imputa los nulos de `lesividad` con cero.

## 📊 Ejercicio 8. Estadísticos básicos

1. ¿Cuál es la calle con más accidentes?
2. ¿Cuáles son los 3 distritos con mayor número de accidentes?
3. ¿Cuál es la distribución de la variable `sexo`?
4. Calcula media, mediana, máximo y mínimo de `lesividad`.

## 📈 Ejercicio 9. Visualización

1. Representa un histograma de la columna `lesividad`.
2. Representa un gráfico de barras con el número de accidentes por distrito.

## 🧠 Ejercicio 10. Reflexión

¿Qué decisiones sobre limpieza y transformación de datos te han parecido más relevantes? ¿Qué errores o inconsistencias podrías haber cometido si no hubieras analizado primero la calidad del dataset?
